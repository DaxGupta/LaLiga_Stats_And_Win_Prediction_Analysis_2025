```python
import pandas as pd

df_2014_2015 = pd.read_html("https://fbref.com/en/comps/12/2014-2015/2014-2015-La-Liga-Stats",
                            attrs={"id" : "results2014-2015121_overall"})[0]
```

```python
df_2014_2015
```

```python
df_2024_2025 = pd.read_html("https://fbref.com/en/comps/12/La-Liga-Stats" ,
                            attrs={"id" : "results2024-2025121_overall"})[0]

print(df_2024_2025.head())
```

```python
# Define the range of seasons
seasons = [(2014, 2015), (2015, 2016), (2016, 2017), (2017,2018), (2018,2019),
 (2019,2020), (2020,2021), (2021,2022), (2022,2023), (2023,2024)]

# Dictionary to store the DataFrames
season_dfs = {}

# Iterate through each season
for start, end in seasons:
    season_str = f"{start}-{end}"
    url = f"https://fbref.com/en/comps/12/{season_str}/{season_str}-La-Liga-Stats"
    table_id = f"results{start}-{end}121_overall"

    try:
        df = pd.read_html(url, attrs={"id": table_id})[0]
        season_dfs[season_str] = df
        print(f"Successfully fetched data for {season_str}")
    except Exception as e:
        print(f"Failed to fetch data for {season_str}: {e}")
```

```python
season_dfs[2024-2025]= df_2024_2025
```

```python
season_dfs[2014-2015].columns
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
for i in season_dfs:
    season_dfs[i].columns = season_dfs[i].columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
season_dfs[2014-2015].columns
```

```python
j=0
for i in season_dfs:
  season_dfs[i]["season"]=2014+j
  j+=1
```

```python
season_dfs[2024-2025].head()
```

```python
full_data = pd.concat(season_dfs.values(), ignore_index=True)
full_data.head()
```

```python
full_data.tail()
```

```python
full_data.to_csv("la_liga_data_2014_2025.csv", index=False)
print("Data saved Successfully !")
```
